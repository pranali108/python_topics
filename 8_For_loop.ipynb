```python
# find out only int type of data 

d = [12,1.5, True, 'Python',70]
d
```

```python
for i in d:
    print(i)
```

```python
for i in d:
    if type(i) == int:
        print(i)
```

```python
for i in d:
    if type(i) == str:
        print(i)
```

```python
for i in d:
    if type(i) == str:
        print(i[::-1])
```
