# Conditional Statements Practice
### 1. Eligibility Check:

```python
# Write a program that takes the age of a person as input and prints whether they are eligible to vote (18 years or older).

age = int(input("Enter Your Age"))
if age >= 18:
    print('You are eligible to vote')
else: 
    print('You are not eligible for vote')
```

### 2.Number Classification:


```python
# Take a number as input and print whether it is positive, negative, or zero.

num = int(input("Enter your no"))
if num > 0:
    print(f'{num} is Positive')
elif num < 0:
    print(f'{num} is Negative')
else:
    print(f'{num} is zero')
```

```python
# 3. Grade Evaluation:
# Write a program that asks for a student's score (out of 100) and prints their grade
# based on the following criteria:
# 90 and above: "A"
# 80-89: "B"
# 70-79: "C"
# 60-69: "D"
# Below 60: "F"

marks = float(input('Enter Your Marks'))
if marks >= 90:
    print('A')
elif marks >= 80:
    print('B') 
elif marks >= 70:
    print('C')
elif marks >= 60:
    print('D')
else:
    print('Fail')
```

```python
# 4. Leap Year Checker:
# Write a program to check if a given year is a leap year. (Hint: A year is a leap
# year if it is divisible by 4 but not by 100, except when it’s also divisible by 400.)

year = int(input("Enter a year: "))
if(year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f'{year} is a leap year')
else:
    print(f"{year} is not a leap year.")
```

```python
# 5. Max of Three Numbers:
# Take three numbers as input and print the largest of the three.

num1 = float(input('Enter 1st number'))
num2 = float(input('Enter 2nd number'))
num3 = float(input('Enter 3rd number'))

if num1 > num2 and num1 > num3:
    print(f'{num1} is largest')
elif num2 > num1 and num2 > num3:
    print(f'{num2} is largest')
else:
    print(f'{num3} is largest')
```

```python
# 6. Even or Odd:
# Write a program that takes an integer input and checks if it is even or odd

num = int(input('Enter Your Number'))
if num % 2 == 0:
    print(f"{num} is Even")
else:
    print(f"{num} is Odd")
```

```python
# 8. Discount Calculation:
# Write a program that calculates the final price of an item after applying a discount
# based on the price:
#  Above $100: 20% discount
#  $50-$100: 10% discount
#  Below $50: No discount

price = float(input('Enter the price of the item'))

if price > 100:
    discount = price * 0.2
    final_price = price - discount
    print(f"You get a 20% discount of ${discount}")
    print(f"Final price: ${final_price}")
elif price >= 50 and price <= 100:
    discount = price * 0.1
    final_price = price - discount
    print(f'You get a 10% discount of ${discount}')
    print(f"Final price: ${final_price}")
else:
    print("No discount available")
    print(f"Final price: ${price}")
```

# Basic For Loop Practice Questions

```python
# 1. Print Numbers from 1 to 10:
# Write a program using a for loop to print numbers from 1 to 10.

for i in range(1,11):
    print(i)
```

```python
# 2.Print Each Character in a String:
# Ask the user to enter a word and use a for loop to print each character in the word

word = str(input("Enter a word: "))

for i in word:
    print(i)
```

```python
#3. Count Up and Down:
# Use a for loop to print numbers from 1 to 5 and then back down to 1.

for i in range(1, 6):
    print(i)

for i in range(4, 0, -1):
    print(i)
```

```python
# 4. Print Elements of a List:
# Define a list of fruits (["apple", "banana", "cherry", "date"]) and use a
# for loop to print each fruit in the list.

fruits = ['apple','banana','cherry','date']

for i in fruits:
    print(i)
```

```python
# 5. Print Even Numbers Between 1 and 10:
# Use a for loop to print all even numbers between 1 and 10.

for i in range(1,11):
    if i % 2 == 0:
        print(i)
```

```python
# 6.Simple Sum of a List:
# Write a program that calculates the sum of numbers in a list ([1, 2, 3, 4, 5])
# using a for loop.

l = [1,2,3,4,5]

sum = 0

for i in l:
    sum = sum + i
print("Sum of list elements:", sum)
```

```python
# 7. Count Down from 5 to 1:
# Use a for loop to print numbers in reverse from 5 to 1.

for i in range(5,0,-1):
    print(i)
```

```python
# 8. Square Each Number in a List:
# Given a list of numbers [1, 2, 3, 4, 5], use a for loop to print the square of
# each number.

num = [1,2,3,4,5]

for i in num:
    print(i * i)
```

```python
# 9. Repeat a Word Multiple Times:
# Ask the user for a word and a number N. Use a for loop to print the word N times

word = input('Enter the word: ')
n = int(input('Enter the number'))

for i in range(n):
    print(word)
```
