### 1. Understanding the Core Concepts Perfectly
Tuples: The Immutable Store
A tuple is an ordered collection of elements, similar to a list, but with one major difference: it is immutable (cannot be changed after creation).

•	Syntax: Uses parentheses ().

•	Key Property: If you try to change an element (e.g., t[0] = 300), Python will throw a TypeError.

•	Methods: Common methods include .count() to find occurrences and .index() to find the position of an element.

Sets: The Unique Collection
A set is an unordered collection of unique elements.

•	Syntax: Uses curly braces {}.
•	Key Property: It automatically removes duplicates. If you create a set with [1, 1, 2], it becomes {1, 2}.

•	Operations: You can add elements with .add(), remove them with .remove(), and perform mathematical operations like Union (all items) and Intersection (common items).

Dictionaries: The Key-Value Pair
A dictionary stores data in key: value pairs, allowing you to look up information efficiently.

•	Syntax: Uses curly braces with colons {key: value}.

•	Key Property: Keys must be unique, and they are used to access the corresponding values (e.g., d['age'] returns 20).

•	Usage: Often used to represent structured data, like a person's profile or a database record.
