### 1. Even or Odd Checker 

```python
# Write a function that takes a number as input and returns whether it is even or odd.

num = int(input('Enter the number'))

def even_odd(num):
    if num%2 == 0:
        print(f'{num} is Even Number')
    else:
        print(f'{num} is Odd Number')

even_odd(num)
```

### 2. Maximum of Three Numbers 

```python
# Create a function that accepts three numbers and returns the largest.

def find_max(a,b,c):
    if a >= b and a >= c:
        print(f"{a} is larger")
    elif b >= a and b >= c:
        print(f"{b} is larger")
    else:
        print(f"{c} is larger")
```

```python
find_max(34,85,52)
```

### 3. Simple Calculator 

```python
# Write a function that accepts two numbers and an operator (+, -, *, /) and performs the operation. 

def cal(a,b,operation):
    if operation == "add":
        print(a + b)
    elif operation == "sub":
        print(a - b)
    elif operation == "mul":
        print(a * b)
    elif operation == "div":
        if b != 0:
            print(a / b)
        else:
            print("Division by zero is not allowed.")

    else:
        print("Invalid operation")
    
```

```python
cal(5,7,"add")
```

### 4. Palindrome Checker

```python
# Write a function to check if a given string is a palindrome (reads the same backward).

def is_palindrome(text):
    if text == text[::-1]:
        print(f"{text} is Palindrome")
    else:
        print(f"{text} is Not Palindrome")
```

```python
is_palindrome("level")
```

```python
is_palindrome("Python")
```

### 5. Count Vowels in a String 

```python
# Write a function that counts the number of vowels in a given string.

def count_vowels(word):
    count = 0
    
    for ch in word:
        if ch.lower() in "aeiou":
            count += 1

    print("Number of vowels:", count)
```

```python
count_vowels("book")
```

### 6. Factorial Calculator 

```python
# Create a function that returns the factorial of a given number. 

def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i

    return result

print(factorial(5))
```

### 7. Fibonacci Sequence Generator 

```python
# Write a function that generates the first N terms of the Fibonacci sequence. 

def fibonacci(n):
    sequence = []
    a, b = 0, 1

    for i in range(n):
        sequence.append(a)
        a, b = b, a + b

    return sequence
```

```python
fibonacci(10)
```

### 8. Prime Number Checker 

```python
# Write a function to check if a given number is a prime number.

def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(is_prime(7))
```

### 9. List Sum and Average 

```python
# Create a function that accepts a list of numbers and returns the sum and average. 

def sum_and_average(numbers):
    total = sum(numbers)
    average = total / len(numbers)

    return total, average

print(sum_and_average([10, 20, 30, 40, 50]))
```

### 10. Find Grade from Percentage 

```python
# Write a function that takes marks of 3 subjects, calculates the average, and returns the 
# grade based on the percentage.

def calculate_grade(mark1, mark2, mark3):
    average = (mark1 + mark2 + mark3) / 3

    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 60:
        grade = "C"
    elif average >= 35:
        grade = "D"
    else:
        grade = "F"

    return average, grade

result = calculate_grade(85, 90, 80)

print(result)
```

### 11. Password Validator 


```python
#  Password Validator 
# Write a function that checks if a password meets the following conditions: 
# o At least 8 characters 
# o Contains uppercase, lowercase, number, and symbol

def validate_password(password):
    if len(password) < 8:
        return "Invalid Password"

    upper = False
    lower = False
    digit = False
    symbol = False

    for ch in password:
        if ch.isupper():
            upper = True
        elif ch.islower():
            lower = True
        elif ch.isdigit():
            digit = True
        else:
            symbol = True

    if upper and lower and digit and symbol:
        return "Valid Password"
    else:
        return "Invalid Password"

print(validate_password("Python@123"))
```

### 12. Number of Words in a Sentence 

```python
# Create a function that counts how many words are in a given sentence.

def count_words(sentence):
    words = sentence.split()
    return len(words)

print(count_words("Python is easy to learn"))
```

### 13. Reverse Each Word 

```python
# Write a function that reverses each word in a sentence while keeping their order. 

def reverse_words(sentence):
    words = sentence.split()

    result = []

    for word in words:
        result.append(word[::-1])

    return " ".join(result)

print(reverse_words("Python is easy"))
```

### 14. Leap Year Checker 

```python
# Create a function that checks whether a year is a leap year or not. 

def leap_year(year):

    if year % 400 == 0:
        return True

    elif year % 100 == 0:
        return False

    elif year % 4 == 0:
        return True

    else:
        return False

print(leap_year(2024))
```
