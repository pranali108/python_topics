```python
# Part B: Nested Lists
# Given: nested = [1, 2, [100, 200, [300, 400]], 5]
```

```python
nested = [1, 2, [100, 200, [300, 400]], 5]

print(nested.index([100, 200, [300, 400]]))
```

```python
#Access the number 200.
nested[2][1]
```

```python
# Access the number 400.
nested[2][2][1]
```

```python
nested
```

```python
nested[2][1] = 200
nested
```

# Part C: Advanced & Mixed Concepts
Given: data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] 

```python
# How do you access the entire second row [4, 5, 6]?
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
data[1][0:]
```

```python
data[1][1:3]
```

```python
# How do you access the number 9 using negative indexing?

data[-1][-1]
```

```python
# If x = [1, 2, 3], what is the difference between x.append([4, 5]) and x.extend([4, 5])?
x = [1, 2, 3]
x.append([4, 5])  # append adds the list as one item. 
x
```

```python
x = [1, 2, 3]
x.extend([4, 5])  # extend adds items individually: [1, 2, 3, 4, 5].
x
```

```python
# Given: z = [10, 20, 50, 76, 89, 56, 87, 90]. What is z[-3::-1]?

z = [10, 20, 50, 76, 89, 56, 87, 90]
z[-3::-1]
```

```python
# Write code to reverse only the middle three elements of a 5-element list.

list = [1,2,3,4,5]
list[1:4][::-1]
```
