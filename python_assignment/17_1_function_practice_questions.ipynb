# Part A: Data Transformation (Level 1)

```python
# Binary Converter: Convert a list of booleans [True, False, True] into integers [1, 0, 1] using comprehension.

boolean_list = [True,False,True]

[int(i) for i in boolean_list]
```

```python
# Pass/Fail: Given marks = [20, 45, 80, 30], create a list that says "Pass" if marks ≥ 35, else "Fail".

marks = [20,45,80,30]

['Pass'if i >= 35 else "Fail" for i in marks]
```

```python
# Positive/Negative: Transform [10, -5, 0, -2] into ['Pos' if i >= 0 else 'Neg' for i in data].

data = [10,-5,0,2]
['Positive' if i >= 0 else "Negative" for i in data]
```

```python
# Case Normalizer: Given a list of strings, convert them to uppercase if they start with 'a', otherwise leave them lowercase.

list_of_string = ["apple",'banana','mango']

[i.upper() if i.startswith('a') else i.lower() for i in list_of_string]
```

```python
# Zero Replacer: In a list of numbers, replace every 0 with the string "Missing".

numbers = [0,1,4,6,3,0,8,10,0]

['Missing' if i == 0 else i for i in numbers]
```

## Part B: Nested & Advanced Transformation (Level 2)

```python
# Multi-Level Grading: (Advanced) Use nested if-else to label numbers as "Small" (<10), "Medium" (10-20), or "Large" (>20).

numbers = [25,14,7]

["small" if i < 10 else "Medium" if i <= 20 else "Large" for i in numbers]
```

```python
# Price Taxer: Apply a 15% tax to prices over 100, and 5% tax to prices below 100.

price = [98,120,35,150]

[p * 1.15 if p > 100 else p * 1.05 for p in price]
```

```python
# Vowel Highlighter: Given a word, create a list where vowels are Uppercase and consonants are Lowercase.

[c.upper() if c.lower() in 'aeiou' else c.lower() for c in "book"]
```

```python
# List to Dictionary Transformation: Create a dictionary from a list where the value is "Square" if the number is even and "Cube" if odd.

{i:'Square' if i % 2 == 0 else "Cube" for i in range(5)}
```

```python
# Length Checker: Transform a list of words into their lengths, but if the word is "secret", replace the length with the string "Hidden".

words = ["apple", "secret", "python", "code"]
[len(i) if i != "secret" else "Hidden" for i in words]
```

### Part C: Mathematical & Logic Operations (Level 3)

```python
# Absolute Values: Implement your own absolute value logic: if i < 0, return -i, else i.

nums = [12,4,0,-2,7]

[-i if i<0 else i for i in nums]
```

```python
# FizzBuzz Lite: Create a list for range(1, 16) where multiples of 3 are replaced by "Fizz", and others stay as numbers.

["Fizz" if i%3 == 0 else i for i in range(1,16)]
```

```python
# Parity Multiplier: Multiply even numbers by 2 and odd numbers by 10.

nums = [1,2,3,4,5,6,7,8,9]

[i*2 if i % 2 == 0 else i*10 for i in nums]
```

```python
# Safe Division: Given a list of divisors, return 100/i if i != 0, else return 0.

divisors = [2, 5, 0, 10, 4]
[100/i if i != 0 else 0 for i in divisors]
```

```python
# Boolean Toggle: Flip a list of bits (0 becomes 1, 1 becomes 0).

[1 if x==0 else 0 for x in [0,1,0,1]]
```

### Part D: Top 1% Practice (Level 4)

```python
# Matrix Identity: Create a 3x3 identity matrix (1s on the diagonal, 0s elsewhere) using nested comprehensions and if-else.

[[1 if row == col else 0 for col in range(3)]for row in range(3)]
```

```python
# Whitespace Cleaner: Given a list of strings, strip them if they are strings, otherwise leave them as is (e.g., if they are ints).

data = ['Hello  ',12,' Hi',54,76,31]
[s.strip() if type(s) == str else s for s in data]
```

```python
# List Flattening with Filter: Flatten List Flattening with Filter: Flatten [[1, 2], [3, 4]] but only if the sub-list sum is greater than 3. but only if the sub-list sum is greater than 3.

Flatten = [[1, 2], [3, 4]]
[i for sub in Flatten if sum(sub)>3 for i in sub]
```

```python
# Conditional Dictionary Swap: Swap keys and values in a dictionary, but only if the value is a string.

d = {
    "a": "apple",
    "b": 10,
    "c": "cat",
    "d": 20
}
{v: k for k, v in d.items() if type(v) == str}
```

```python
# Function Mapping: Create a list that applies math.sqrt(i) if i > 0 else returns 0 for a range of numbers.

import math
numbers = range(-3, 6)

result = [math.sqrt(i) if i > 0 else 0 for i in numbers]

print(result)
```
