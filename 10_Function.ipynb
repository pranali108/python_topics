```python
# similar code for nested list, and you have to find out int type of data from list

# [1,1.5,True,[11,15.2,False], (12,12.6,'Python'), [10,20,10.5,[1,2,3,4,'Hello']]

def get_int_values(*args):
    result = []

    for i in args:
        if type(i) == int:
            result.append(i)

        elif type(i) == list or type(i) == tuple:
            result.extend(get_int_values(*i))

    return result
```

```python
get_int_values(
1,1.5,True,[11,15.2,False], (12,12.6,'Python'), [10,20,10.5,[1,2,3,4,'Hello']])
```
