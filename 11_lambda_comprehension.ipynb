## Comprehension

### List Comprehension

```python
for i in range(1,6):
    print(i)
```

```python
[i for i in range(1,6)]
```

```python
[i+1 for i in range(1,11)]
```

```python
[i*i for i in range(1,11)]
```

```python
for i in range(1,11):
    if i%2 == 0:
        print(i)
```

```python
[i for i in range(1,11) if i%2 == 0]
```

```python
a = (i for i in range(1,11))
```

```python
a
```

```python
for i in a:
    print(i)
```

```python
# set
```

```python
{i for i in range(1,11)}
```

```python
{'even' for i in range(1,11) if i%2 == 0}
```

```python
# Dict
```

```python
{i:i for i in range(1,11)}
```

```python
{i:i+10 for i in range(1,11)}
```

```python
{i:i*i for i in range(1,11)}
```
