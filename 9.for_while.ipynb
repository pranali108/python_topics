```python
p = [1, 1.5, 'hello', [2,[3,3.5,'hello'], 2.5, 'python'], True]
p
```

```python
for i in p:
    if type(i) == int:
        print(i)
    elif type(i) == list:
        for ele in i:
            if type(ele) == int:
                print(ele)
            elif type(ele) == list:
                for j in ele:
                    if type(j) == int:
                        print(j)
```

# While loop

```python
age = 15
while age < 18:
    print('Student', age)
    age = age + 1
```

```python
time = 0
while time <= 10:
    print(time)
    time = time + 1
```

```python
button = 'hello'

while button != 'exit':
    print(button)
    button = input("Enter 'exit' to stop loop")
```

# Break

```python
for i in range(1,6):
    if i == 3:
        break
    else:
        print(i)
```

```python
for i in 'hello':
    if i == 'l':
        break
    print(i)
```

# Continue

```python
for i in range(1,6):
    if i == 3:
        continue
    print(i)
    print('Hello')
```

```python
for i in 'hello':
    if i == 'l':
        continue
    print(i,'python')
    print(i,'python')
    print(i,'python')
    print(i,'python')
```

```python
roll = [1,2,3,4,5,6]
```

```python
for i in roll:
    if i== 1:
        continue
    elif i == 4:
        continue
    print(f'Homework 10 times for roll no {i}')
```
